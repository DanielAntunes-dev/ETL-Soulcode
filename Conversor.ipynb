# LER O PDF A PARTIR DE UMA URL



```python
def pdf_para_dataframe(url):
    # Faz o download do PDF a partir do URL
    resposta = requests.get(url)
    with open("temp.pdf", "wb") as f:
        f.write(resposta.content)

    # Extrai tabelas do arquivo PDF
    tabelas = tabula.read_pdf("temp.pdf", pages='all')

    # Inicializa um DataFrame vazio para armazenar as tabelas combinadas
    dataframe_combinado = pd.DataFrame()

    # Combina todas as tabelas em um único DataFrame
    for tabela in tabelas:
        df = tabela.copy()
        dataframe_combinado = pd.concat([dataframe_combinado, df], ignore_index=True)

    # Retorna o DataFrame resultante
    return dataframe_combinado

# URL do arquivo PDF que você deseja converter
url_do_pdf = 'https://www.ibam-concursos.org.br/documento/161/cp/006.pdf'

# Chama a função para converter o PDF em um DataFrame único
dataframe_combinado = pdf_para_dataframe(url_do_pdf)
```

```python
def pdf_para_dataframe(ruta_arquivo):
    # Extrai tabelas do arquivo PDF
    tabelas = tabula.read_pdf(ruta_arquivo, pages='all')

    # Inicializa um DataFrame vazio para armazenar as tabelas combinadas
    dataframe_combinado = pd.DataFrame()

    # Combina todas as tabelas em um único DataFrame
    for tabela in tabelas:
        df = tabela.copy()
        dataframe_combinado = pd.concat([dataframe_combinado, df], ignore_index=True)

    return dataframe_combinado

# Rota do arquivo PDF no seu computador
ruta_do_pdf = '/content/pdf_prova.pdf'

# Chama a função para converter o PDF em um DataFrame único
dataframe_combinado2 = pdf_para_dataframe(ruta_do_pdf)
dataframe_combinado2
```

```python
#Remover as colunas innecesarias
coluna_remover = ['Clase','Discursiva','Titulos']
df3 = df3.drop(coluna_remover,axis=1)
df3
```
